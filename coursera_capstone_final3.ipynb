```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
def URL_get(info):
    results = requests.get(info).json()
    venues = results['response']['venues']
    dataframe = json_normalize(venues)
    filtered_columns = ['name', 'categories'] + [col for col in dataframe.columns if col.startswith('location.')] + ['id']
    dataframe_filtered = dataframe.loc[:, filtered_columns]
    dataframe_filtered['categories'] = dataframe_filtered.apply(get_category_type, axis=1)
    dataframe_filtered.columns = [column.split('.')[-1] for column in dataframe_filtered.columns]
    return dataframe_filtered
```

```python
column_names = ['Latitude','Longitude','Name','Address','Brand','ID','URL']
df_loc_big = pd.DataFrame(columns = column_names)
df_loc_big
```

```python
for i,j in enumerate(df_loc['URL']):
    getting = URL_get(j)
    for lat, lng, name,id in zip(getting.lat, getting.lng, getting.name,getting.id):
        df_loc_big['Address'].append(df_loc['Address'].iloc[i])
        df_loc_big['Brand'].append(df_loc['Brand'].iloc[i])
        df_loc_big['URL'].append(df_loc['URL'].iloc[i])
        df_loc_big['Latitude'].append(lat)
        df_loc_big['Longitude'].append(lng)
        df_loc_big['Name'].append(name)
        df_loc_big['ID'].append(id)
```
